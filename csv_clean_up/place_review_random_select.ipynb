# 景點,隨機選30個csv裡的隨機10則評論

```python
import pandas as pd
import numpy as np
import re ,os,random
```

```python
# 選30個檔案,每個檔案選10筆
file = os.listdir("Data/raw_data/attraction/review/review_clean/")
# 隨機抽取30個整數,return nparray
random_ = np.random.choice(len(file),30,replace=False)
```

```python
# [\u0800-\u4e00] 日語
# [\uAC00-\uD7A3] 韓語
folderPath = "random_review_text_attraction"
listReview=[]
for i in random_:
    df = pd.read_csv(f"Data/raw_data/attraction/review/review_clean/{file[i]}")
    # 隨機抽取各個檔案裡的10則評論
    a = df.loc[:,"review_text"].sample(n=10,random_state=1)
    # 每則評論都要過濾不重要的文字 ex: 日韓英語...
    for j in range(10):    
        b = a.index[j]
        regex = r"\\|\n|[a-zA-Z]|\(原始評論\)|\.|\’|[\u0800-\u4e00]|\(由 Google 提供翻譯\)|[\uAC00-\uD7A3]"
        strResult = re.sub(regex, "", a[b])
        listReview.append({
            "review_text":strResult
        })
df = pd.DataFrame(listReview)
df.to_csv(f'Data/raw_data/attraction/review/{folderPath}.csv',mode='a', header=False, index=False)
```
