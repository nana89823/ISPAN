```python
def visit():
    
    try:
        url = "https://www.thsrc.com.tw/"
    
        driver.get(url)
        
        sleep(1)   
        
        alert = driver.find_element(
            By.CSS_SELECTOR,
            "button.swal2-confirm.swal2-styled"
        ).click()
        
    except TimeoutException:
        print("逾時")
```

```python
def setDropDownMenu():
    # 直接選網路訂票功能
    online_Booking = driver.find_element(
        By.CSS_SELECTOR,
        "a[id='a_irs']"
    ).click()
    sleep(3)
    
    # 取得網路訂票功能的iframe
    iframe = driver.find_element(
        By.CSS_SELECTOR,
        "iframe[id='irsUrl']"
    )
    
    # 切換到 iframe 當中
    driver.switch_to.frame(iframe)
    
    # 運用lxml剖析網站元素
    html = driver.page_source
    soup = bs(html,"lxml")
    
    # 起站點
    print(soup.select("select.form-control.mt-0.select-type01")[0].get_text())    # 列出目前的全部的站點
    Station = input("請輸入起站點:")
    start_Station = Select(driver.find_element(
        By.CSS_SELECTOR,
        "select[id='select_location01']"
    ))    
    start_Station.select_by_visible_text(Station)
    
    # 終站點
    print(soup.select("select.form-control.mt-0.select-type01")[1].get_text())    # 列出目前的全部的站點
    Station = input("請輸入終站點:")
    end_Station = Select(driver.find_element(
        By.CSS_SELECTOR,
        "select[id='select_location02']"
    ))
    end_Station.select_by_visible_text(Station)
    
    # 單程　(0代表單程票，1代表來回票)
    way = input("請輸入'單程'or '去回程':")
    way_ = Select(driver.find_element(
        By.CSS_SELECTOR,
        "select[id='typesofticket']"
    ))
    way_.select_by_visible_text(way)
    
    # 選擇去程日期
    latest_Date = driver.find_element(
        By.CSS_SELECTOR,
        "input.form_date.form-control"
    )
    print(f'最晚只能訂到:{latest_Date.get_dom_attribute("slimit")}')
    go_date = (input("去程日期")) 
    driver.execute_script("document.getElementById('Departdate02').value ="+go_date)
    
    # 選擇去程時刻
    go_Time_H = input("請輸入去程時間(幾點):")
    go_Time_M = input("請輸入去程時間(幾分 0 or 30):")
    driver.execute_script("document.getElementById('toPortalTimeTable').value =" + go_Time_H + go_Time_M)
    
    
    
    sleep(2)
```
