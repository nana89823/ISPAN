```python
import requests as req 
import os,json,re
from bs4 import BeautifulSoup as bs
from fake_useragent import UserAgent 
```

```python
ua = UserAgent(cache= True)
url = "https://www.gutenberg.org/browse/languages/zh"
res = req.get(url , headers={"user-agent":ua.random})
# print(res.status_code)
soup = bs(res.text,"lxml")


# 新增古騰堡計劃資料夾

folderPath ="古騰堡計劃_前兩百本電子書"
if os.path.exists(folderPath) == False:
    os.makedirs(folderPath)
```

```python
# 找出前300本書的網址，並加到list_title_herf裡

def search_title_href():
    ResultSetElement = soup.select("li.pgdbetext a" ,limit=300)
    list_title_href = []
    for index,a in enumerate(ResultSetElement):
        list_title_href.append("https://www.gutenberg.org/"+a["href"])
    return index,list_title_href
```

```python
# 找出前300本的書名並加到list_title

def search_title():
    ResultSetElement = soup.select("li.pgdbetext a",limit=300)
    list_title =[]
    for index,a in enumerate(ResultSetElement):
        regex = re.sub(r'\W'," ",a.get_text())
        list_title.append(regex)
    return index, list_title
```

```python
# 找出開啟書內容的網址，並加到list_content_href裡

def search_content_href(list_title_href):
    list_content_href = []
    for index,link in enumerate(list_title_href):
        res = req.get(link, headers={"user-agent":ua.random})
        soup = bs(res.text,"lxml")
        
        #判斷是英文書還是中文書,兩者抓取得html不同
        ResultSetElement = soup.select("tr.even")[3].get("about")
        if ResultSetElement == None:
            ResultSetElement = soup.select("tr.odd")[3].get("about")
            list_content_href.append(ResultSetElement)
        else:
            list_content_href.append(ResultSetElement)
    return index,list_content_href
```

```python
# 抓取內文並寫入檔案

def write_to_text(list_content_href):
    i = 0
    title = search_title()[1]
    for link in list_content_href:
        res = req.get(link, headers={"user-agent":ua.random})
        res.encoding="utf-8"
        soup = bs(res.text,"lxml")
        
        # 建立電子書的text檔案
        path = "./古騰堡計劃_前兩百本電子書/"+title[i]+".text"
        f = open(path , "a", encoding="utf-8")
    
        # 篩選英文書與中文書並分別寫入對應的檔案內
        ResultSetElement = soup.select_one("body").get_text()
        regex = r'(Title[\u4e00-\u9fa5_\a-zA-Z0-9\S]{20})([a-zA-Z\W\d]+)([\u4e00-\u9fa5\S\s]+)(?=End)'
        search = re.search(regex,ResultSetElement)
        if search == None :
            regex = r'[a-zA-Z\d\W]+([\u4e00-\u9fa5\S\s]+)?'
            search = re.search(regex,ResultSetElement)
            f.write(search[0])
            f.close()
        else:
            f.write(search[1])
            f.write(search[3])
            f.close()
        i+=1
        print(f'電子書第{i}本:{title[i-1]},已寫入成功')
```

```python
if __name__ == "__main__":
    get_title_href = search_title_href()[1]
    get_content_href = search_content_href(get_title_href)[1]
    write_to_text(get_content_href)
```
