```python
import pandas as pd
import jieba
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def get_custom_stopwords(stop_words_file):
    with open(stop_words_file,encoding="utf-8") as f:
        stopwords = f.read()
    stopwords_list = stopwords.split('\n')
    custom_stopwords_list = [i for i in stopwords_list]
    return custom_stopwords_list
```

```python
df = pd.read_csv('Data/raw_data/restaurant/review/review_label_restaurant - 工作表2.csv',header=None)
df = df.iloc[1:,1:3]
df
```

```python
X = df[[2]]
y = df[1]
print(X.shape)
print(y.shape)
```

```python
stop_words_file = "stopwords.txt"
stopwords = get_custom_stopwords(stop_words_file)
```

```python
 X[2]
```

```python
poun=('，','。','！','：','「','」','…','、','？','【','】','.',':','?',';','!','~','`','+','-','<','>','/','[',']','{','}',"'",'"', '\n'," ","")
list_ = []
jieba.set_dictionary('dict.txt.big')
for originReview in  X[2]:
    # 刪除標點符號
    filteredText = "".join(c for c in originReview if c not in (poun))
    # 讀入停用詞
    stopWords = {}.fromkeys([ line.rstrip() for line in stopwords])
    #  以精確模式分詞
    words = jieba.cut(filteredText, cut_all=False)
    # 去除停用詞後, 以空白斷詞
    final = ''
    for seg in words:
        if seg not in stopWords:
            seg+=' '
            final += seg
    list_.append(final)
X["cutted_comment"] = list_
```

```python
X["cutted_comment"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1,test_size=0.2)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
vector = CountVectorizer()
term_matrix = pd.DataFrame(vector.fit_transform(X_train.cutted_comment).toarray(), columns=vector.get_feature_names_out())
term_matrix.head()
```

```python
max_df = 0.8 # 出現超過0.8比例的內容刪掉(過於頻繁)
min_df = 3   # 低於3次的關鍵次刪掉(過少,獨特)
vector = CountVectorizer(max_df = max_df, 
                       min_df = min_df, 
                       token_pattern=u'(?u)\\b[^\\d\\W]\\w+\\b',  # 刪除數字
                       )
```

```python
term_matrix = pd.DataFrame(vector.fit_transform(X_train.cutted_comment).toarray(), columns=vector.get_feature_names_out())
```

```python
term_matrix
```

# Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB(alpha=2)

from sklearn.pipeline import make_pipeline
pipe = make_pipeline(vector, nb)

from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train.cutted_comment, y_train, cv=5, scoring='accuracy').mean()
```

```python
pipe.fit(X_train.cutted_comment, y_train)
```

```python
y_pred = pipe.predict(X_test.cutted_comment)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)

con_matrix = confusion_matrix(y_test, y_pred)

num_correct_samples = accuracy_score(y_test, y_pred, normalize=False)

print('accuracy: {}'.format(accuracy))

print('confusion matrix: {}'.format(con_matrix))

print('number of correct sample: {}'.format(num_correct_samples))
```
